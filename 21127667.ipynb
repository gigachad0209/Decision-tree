```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import tree
import seaborn as sns
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz


```

```python
whole_set = pd.read_csv("/content/poker-hand-data.csv", header=None)
whole_set.columns = ['S1', 'C1','S2', 'C2','S3', 'C3','S4', 'C4','S5', 'C5','Label']
feature = whole_set.loc[:,whole_set.columns != 'Label']
label = whole_set['Label']
```

```python
feature_train, feature_test, label_train, label_test = train_test_split(feature, label, train_size=0.4, test_size=0.6, stratify=label)
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth = 3)
clf= clf.fit(feature_train, label_train)

tree_data = export_graphviz(clf)
graph = graphviz.Source(tree_data)
display(graph)
```

```python
label_hat = clf.fit(feature_train, label_train).predict(feature_test)
print()
print(classification_report(label_test, label_hat))
print()
ConfusionMatrixDisplay.from_predictions(label_test, label_hat, cmap='Reds')
```
